```python
import pandas as pd
import numpy as np
import glob
```

```python
# Read files in data folder to a dataframe
path ='data' # use your path
allFiles = glob.glob(path + "/*.csv")
loans_df = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,skiprows=1,low_memory=False)
    list_.append(df)
loans_df = pd.concat(list_)
```

```python
loans_df.head(5)
```

```python
loans_df.info()
```

There are total of 1321864 rows and 151 columns in the dataset

```python
#Now lets see how many null values are present in each column
loans_df.isnull().sum().sort_values(ascending=False)
```

```python
#From the above you can see there are many missing values in each colum/attribute.
#Now we can get rid of the columns which have more than 30% missing or null values
threshold_count = len(loans_df)*0.7 
# Drop any column with more than 30% missing values
filtered_loans_df = loans_df.dropna(thresh=threshold_count,axis=1)
```

```python
filtered_loans_df.info()
```

After removing the columns that have more than 30% missing columns, we have total number rows of 1321864. but the columns have reduced to 93.

```python
#Now lets see how many null values are present in each column
filtered_loans_df.isnull().sum().sort_values(ascending=False)
```

```python
filtered_loans_df.info()
```

After filtering the dataset we have 93 columns. Among 93 columns, 67 are of numerical data type and 26 are categorical data types

```python
# Now we will drop some columns which doesn't really affect our analysis
#The cloulms are url,id 
filtered_loans_df = filtered_loans_df.drop(['id' ,'url'],axis=1)
```

```python
#Drop the rows which has null values in each column
filtered_loans_df = filtered_loans_df.dropna(axis=0,how='all')
```

```python
filtered_loans_df.info()
```

Now the column count has been reduced to 91 and row count also reduced to 1321847

```python
# Now we will remove the columns that have only one value
filtered_loans_df = filtered_loans_df.loc[:,filtered_loans_df.apply(pd.Series.nunique) != 1]
```

```python
filtered_loans_df.info()
```

Now the number of columns are reduced to 90

```python
#Now filter the columns that have less than 4 unique values
for col in filtered_loans_df.columns:
    if (len(filtered_loans_df[col].unique()) < 4):
        print(filtered_loans_df[col].value_counts())
        print()
```

 There are total 7 columns that have only 2 unique values.
 <p> "term" is an important factor /attribute in any loan. "debt_settlement_flag" can be important to our analysis.so we don't remove term and debt_settlement_flag from the dataset</p>
 <p>But we can drop the rest of the columns </p>


```python
filtered_loans_df = filtered_loans_df.drop(['pymnt_plan','initial_list_status','hardship_flag','disbursement_method' ,'application_type'],axis=1)
```

```python
filtered_loans_df.info()
```

```python
# Now we will divide our dataset into 2 dataframes. 
#1. Numerical 2. Categorical
filtered_categorical_loans_df = filtered_loans_df.select_dtypes([object])
filtered_numeric_loans_df = filtered_loans_df.select_dtypes(['float64','int64'])
```

```python
filtered_categorical_loans_df.head(5)
```

```python
# Print unique values for purpose and title
filtered_categorical_loans_df.purpose.value_counts()
filtered_categorical_loans_df.title.value_counts()
```

From the above you can see the title and purpose have similar values.
<p>'title' is the description of the purpose of the loan. So we can drop the tilte.</p>

<p>Similarly, zip_code and addr_state gives the same infomration. so we can drop zip_code</p>

<p>Also, subgrade is the sub category of grade.So we will drop subgrade column too


```python
filtered_categorical_loans_df = filtered_categorical_loans_df.drop(['title','zip_code','sub_grade'],axis=1)
```

```python
filtered_categorical_loans_df.info()
```

```python
#Now replace the missing values with mode
filtered_categorical_loans_df = filtered_categorical_loans_df.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
#Now the number of missing values in each column should be zero
filtered_categorical_loans_df.isnull().sum().sort_values(ascending=False)
```

```python
filtered_categorical_loans_df.info()
```

```python
filtered_categorical_loans_df.head()
```

```python
#The colums int_rate and revol_util are in percantage. we will convert it to numeric values.
filtered_categorical_loans_df['int_rate'] = filtered_categorical_loans_df['int_rate'].replace('%','',regex=True).astype('float')

filtered_categorical_loans_df['revol_util'] = filtered_categorical_loans_df['revol_util'].replace('%','',regex=True).astype('float')
```

```python
filtered_categorical_loans_df.head()
```

The columns issue_d ,earliest_cr_line,last_pymnt_d,last_credit_pull_d are in Mon-yyyy date format.We are going to create 2 numeric columns from each date by splitting into a month and year.

```python
filtered_categorical_loans_df['issue_year'] = pd.DatetimeIndex(filtered_categorical_loans_df['issue_d']).year
filtered_categorical_loans_df['issue_month'] = pd.DatetimeIndex(filtered_categorical_loans_df['issue_d']).month
```

```python
filtered_categorical_loans_df['earliest_cr_line_month'] = pd.DatetimeIndex(filtered_categorical_loans_df['earliest_cr_line']).month
filtered_categorical_loans_df['earliest_cr_line_year'] = pd.DatetimeIndex(filtered_categorical_loans_df['earliest_cr_line']).year
```

```python
filtered_categorical_loans_df['last_pymnt_month'] = pd.DatetimeIndex(filtered_categorical_loans_df['last_pymnt_d']).month
filtered_categorical_loans_df['last_pymnt_year'] = pd.DatetimeIndex(filtered_categorical_loans_df['last_pymnt_d']).year
```

```python
filtered_categorical_loans_df['last_credit_pull_month'] = pd.DatetimeIndex(filtered_categorical_loans_df['last_credit_pull_d']).month
filtered_categorical_loans_df['last_credit_pull_year'] = pd.DatetimeIndex(filtered_categorical_loans_df['last_credit_pull_d']).year
```

```python
#lets drop issue_d ,earliest_cr_line,last_pymnt_d,last_credit_pull_d
filtered_categorical_loans_df = filtered_categorical_loans_df.drop(['issue_d','earliest_cr_line','last_pymnt_d','last_credit_pull_d'],axis=1)
```

```python
filtered_categorical_loans_df['term'].value_counts()
```

```python
#Convert the term column to numeruc by removing the months.
filtered_categorical_loans_df['term'] = filtered_categorical_loans_df['term'].replace('months','',regex=True).astype('int')
```

```python
filtered_categorical_loans_df.emp_title.value_counts()
```

```python
#There are total of 381900 unique values for emp_title. For now we can drop emp_title for our analysis
filtered_categorical_loans_df = filtered_categorical_loans_df.drop(['emp_title'],axis=1)
```

```python
filtered_categorical_loans_df.info()
```

```python
#Now create a 2 dataframes from filtered_categorical_loans_df
#1.With categorical values - Stage1 (This is used for data exploration)
#2.Converting categorical values to numerical values - Stage 2
filtered_categorical_loans_stage1_df = filtered_categorical_loans_df.copy(deep=True)
filtered_categorical_loans_stage2_df = filtered_categorical_loans_df.copy(deep=True)
```

```python
filtered_categorical_loans_stage1_df.info()
```

```python
filtered_categorical_loans_stage2_df['loan_status'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
filtered_categorical_loans_stage2_df['loan_status'] = le.fit_transform(filtered_categorical_loans_df['loan_status'])
```

```python
filtered_categorical_loans_stage2_df['loan_status'].value_counts()
```

```python
filtered_categorical_loans_stage2_df = pd.get_dummies(filtered_categorical_loans_stage2_df,drop_first=True)
```

```python
filtered_categorical_loans_stage2_df.head()
```

```python
filtered_categorical_loans_stage1_df.head()
```

```python
#Numeric columns
filtered_numeric_loans_df.info()
```

```python
filtered_numeric_loans_df.isnull().sum().sort_values(ascending=False)
```

```python
#Fill the missing values in each column with mean value
filtered_numeric_loans_df = filtered_numeric_loans_df.apply(lambda x: x.fillna(x.mean()),axis=1)
```

```python
#The number of missing values in each columm should be zero
filtered_numeric_loans_df.isnull().sum().sort_values(ascending=False)
```

```python
#There are four columns that has fico scores
#last_fico_range_low,last_fico_range_high,fico_range_low,fico_range_high
#Create a new column with average of all 4 columns.
filtered_numeric_loans_df['fico_average'] = (filtered_numeric_loans_df['fico_range_high'] + filtered_numeric_loans_df['fico_range_low']) / 2
```

```python
#drop the four fico score columns
filtered_numeric_loans_df = filtered_numeric_loans_df.drop(['fico_range_high','fico_range_low','last_fico_range_high','last_fico_range_low'],axis=1)
```

```python
filtered_numeric_loans_df.info()
```

```python
final_filtered_stage1_df = pd.concat([filtered_numeric_loans_df, filtered_categorical_loans_stage1_df], axis=1)

final_filtered_stage1_df.to_csv('preprocessed_loans_stage1.csv')
```
